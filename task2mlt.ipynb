```python
import matplotlib.pyplot as plt

def visualize_find_s_algorithm(positive_examples):
    hypothesis = ['0'] * len(positive_examples[0])
    visualization = []

    for idx, example in enumerate(positive_examples, start=1):
        for i in range(len(example)):
            if hypothesis[i] == '0':
                hypothesis[i] = example[i]
            elif hypothesis[i] != example[i]:
                hypothesis[i] = '?'
        visualization.append((idx, hypothesis.copy()))

    return visualization

positive_examples_find_s = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Cool', 'Change'],
    ['Sunny', 'Hot', 'High', 'Strong', 'Cool', 'Change']
]

hypothesis_find_s_visualization = visualize_find_s_algorithm(positive_examples_find_s)

# Choose the attribute index to plot (0-based)
attribute_index = 0

# Extract values of this attribute across examples
attribute_values = [hyp[attribute_index] for _, hyp in hypothesis_find_s_visualization]

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(attribute_values) + 1), attribute_values, marker='o', linestyle='-')
plt.title(f'FIND-S Algorithm: Evolution of Attribute {attribute_index + 1}')
plt.xlabel('Example Index')
plt.ylabel('Hypothesis Attribute Value')
plt.xticks(range(1, len(attribute_values) + 1))
plt.grid(True)

plt.savefig('FIND-S_Single_Attribute_Visualization.png', dpi=300)
plt.show()
```
